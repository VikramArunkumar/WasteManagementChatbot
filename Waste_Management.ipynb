```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#import seaborn as sb
```

```python
filename = 'https://raw.githubusercontent.com/VikramArunkumar/dataset/refs/heads/main/Waste_national_2018_v1.3.2_9b1bb41.csv'
waste_df = pd.read_csv(filename)
waste_df.head()
```

```python
print("Welcome to the Waste Management Bot!")
name = input("Enter your name: ")

again = True
while again == True:
    user_input = pyip.inputMenu(["What waste management do you use",
"Description of a Waste Management Study and Dataset",
"Some descriptive stats about the national dataset",
"Some data visulization about the national dataset",
"What waste management system you can use to improve San Jose"],
                           numbered = True,
                           prompt = f"Hello {name}, what would you like to\
learn about waste management? \n")
    user_input = user_input.lower()
    if user_input == "what waste management do you use":
        manage = input("Enter a waste management system: ")
        print(f"{manage} is a great way to recycle items! \
Always remember to properly recycle!")
    elif user_input == "description of a waste management study and dataset":
        print("""This dataset is a national dataset that tracks what amount of each
        waste is thrown away (in kg).  This Flow-By-Sector (FBS) dataset
        quantifies waste management for U.S. generated waste by waste management
         pathway and 13 waste materials. The waste management pathways are
         assigned sector codes based on North American Industry Classification
         System (NAICS) 2012 codes.""")

    elif user_input == "some descriptive stats about the national dataset":
        stat = pyip.inputMenu(["What is the maximum flow amount",
                                "What is the minimum flow amount",
                                "How many unique items are there"],
                           numbered = True,
                           prompt = "What do you want to know? \n")
        stat = stat.lower()
        if stat == "what is the maximum flow amount":
            max_row = waste_df.loc[waste_df['FlowAmount'].idxmax(), ['Flowable', 'FlowAmount']]
            print(f"Flowable with maximum FlowAmount: {max_row['Flowable']} "
            f"with Flow Amount: {max_row['FlowAmount']}")

        elif stat == "what is the minimum flow amount":
            min_row = waste_df.loc[waste_df['FlowAmount'].idxmin(), ['Flowable', 'FlowAmount']]
            print(f"Flowable with minimum FlowAmount: {min_row['Flowable']} "
            f"with Flow Amount: {min_row['FlowAmount']}")
        elif stat == "how many unique items are there":
            display(waste_df['Flowable'].nunique())
    elif user_input == "some data visulization about the national dataset":
        flow_data = waste_df[["Flowable", "FlowAmount"]]
        flow_data = flow_data.dropna()
        flow_summary = flow_data.groupby("Flowable")["FlowAmount"].sum().sort_values(ascending=False)
        flow_summary_top_10 = flow_summary.head(10).reset_index()
        plt.figure(figsize=(14,7))
        sns.barplot(data=flow_summary_top_10, x='Flowable', y='FlowAmount')
        plt.title(f"Top 10 Flowables by Total FlowAmount", fontsize=16)
        plt.xlabel('Flowable', fontsize=14)
        plt.ylabel('Total FlowAmount', fontsize=14)
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()

    elif user_input == "what waste management system you can use to improve san jose":
        system = {
            "recycling bin" : ["plastic", "paper", "metal", "glass" , "cardboard"],
            "garbage bin" : ["clothing", "organics", "receipts", "tools", "animal waste"],
            "yard trimmings" : ["branches", "leaf clippings", "sod", "christmas trees"],
            "crv cash back": ["beverage containers", "food packaging"],
            "junk pick-up" : ["big plastic items", "metal items", "electronics",
"carpets", "mattress", "toilets", "furniture", "kitchen appliances"],
            "plastic bag drop off" : ["food packaging", "general packaging",
                                      "padded envelopes"],
            "e-waste" : ["calculators", "phones", "monitors", "televisions",
"printers", "laptops", "headsets", "game consoles", "tablets", "microwaves",
                    "cordless appliances"],
            "hazardous waste" : ["hand sanitizers", "lighters", "cans",
"small tanks", "lightbulbs", "watches", "batteries", "cosmetics", "paint", "fuel tanks"],
            "landfill" : ["portable pools", " scooters", "painted wood", "sandbags", "sinks"]
        }
        use = pyip.inputMenu(["recycling bin", "garbage bin", "yard trimmings",
"crv cash back", "junk pick-up" , "plastic bag drop off" , "e-waste",
                              "hazardous waste" , "landfill" ],
                numbered = True)
        use = use.lower()
        if use in system:
            print(f"What can go in {use} is {system[use]}")

    restart = pyip.inputMenu(["Restart", "Quit"], numbered = True)
    restart = restart.lower()
    if restart == "restart":
        print("Starting again")
    elif restart == "quit":
        again = False
print("Goodbye")
```
